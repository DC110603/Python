Opening a File:::
To open a file, we can use open() function, which requires file-path and mode as arguments.


```python
f=open("example.txt","r")
print(f) # It Returns a File Object
```

Closing a File:::
file.close() method closes the file and releases the system resources. If the file was opened in write or append mode, closing ensures that all changes are properly saved.

```python
#f.close()
print(f.closed) #To check whether the file is closed or not, It returns boolean value
```

Checking File Properties:::
Once the file is open, we can check some of its properties:

```python
print("Filename:", f.name)
print("Mode:", f.mode)
print("Is Closed?", f.closed)
```

Reading a File:::
Reading a file can be achieved by file.read() which reads the entire content of the file.

```python
content = f.read()
print(content)
print(type(content)) #returns the entire data in the file in a string format
```

Writing a File::::
Writing to a file is done using the mode "w". This creates a new file if it doesn’t exist, or overwrites the existing file if it does. The write() method is used to add content.

```python
f=open("example.txt", "w")
f.write("Hello, Python!\n")
f.write("File handling is easy with Python.")

print("File written successfully")
f.close()
```
