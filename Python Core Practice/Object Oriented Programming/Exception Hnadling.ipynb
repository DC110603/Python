.Create a BankAccount class that stores:
• account number
• balance (should not be directly modifiable)
You must:
1. Make the balance attribute inaccessible from outside.
2. Provide functions to deposit/withdraw that validate the amount.
3. Prevent withdrawal if balance becomes negative.
4. Show what happens if someone tries to modify balance directly and why
encapsulation prevents it.

```python
class BankAccount:
    min_balance=1000
    def __init__(self,ac_no,balance,passw):
        self.account_number=ac_no
        self.__balance=balance
        self.__passw=passw
    def deposit(self,amount):
        self.__balance+=amount
    def withdraw(self,withdraw_amt,passw):
        if self.__balance>BankAccount.min_balance and passw == 1234:
            self.__balance-=withdraw_amt
        else:
            return "You Dont Have Minimum Balance in your Account"
    @property
    def show_balance(self,passw):
        if passw == self.__passw:
            return self.__balance
        else:
            return "You have made an Un-Authorised Access"
obj=BankAccount(69696969,10000)
obj.deposit(1000)
print(f" Balance After Deposit: {obj.show_balance(1234)}")
obj.withdraw(2000,1234)
print(f" Balance After Withdraw: {obj.show_balance(1234)}")
obj.withdraw(9000,1234)
print(f" Balance After Withdraw: {obj.show_balance(1234)}")
```
