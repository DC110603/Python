• Create multi-level inheritance with classes A → B → C, each having a method
display() printing the class name. Create object of C and call display(),
showing method resolution.

```python
class A:
    def display(self):
        print("A class Method Displaying")
class B(A):
    def display(self):
        super().display()
        print("B Class Method Displaying")
class C(B):
    def display(self):
        super().display()
        print("C Class Method Displayed")
obj=C()
obj.display()
```

• Implement hierarchical inheritance using a base class Vehicle and two child
classes Car and Bike, each defining a method wheels().

```python
class Vehicle:
    def wheels(self):
        print("wheels of the vehicle")
class Car(Vehicle):
    def wheels(self):
        super().wheels()
        print("Car has 4 Wheels")
class Bike(Vehicle):
    def wheels(self):
        super().wheels()
        print("Bike has 2 wheels")
obj=Bike()
obj1=Car()
obj.wheels()
obj1.wheels()
```

• Create class Employee with an instance method salary(). Create class
Manager(Employee) that overrides salary() and adds an incentive. Demonstrate
both outputs.

```python
class Employee:
    def salary(self):
        return 30000
class Manager(Employee):
    def salary(self):
        incentive=10000
        return super().salary() + incentive

obj=Manager()
obj1=Employee()
print(f"Employee's Base Salary:",obj1.salary())
print(f"Manager's Salary after Incentive:",obj.salary())
```

• Create class MathOps with a static method add(a, b). Create class
AdvancedOps(MathOps) and use the static method without overriding it.

```python
class MathOps:
    @staticmethod
    def addi(a,b):
        return a+b
class AdvancedMathOps(MathOps):
    def swadeep(self,a,b):
        return (super().addi(a,b))
obj=AdvancedMathOps()
print(obj.swadeep(1,5))
```

• Create two classes Father and Mother, both defining a method skills(). Create 
class Child(Father, Mother) and check which skills() runs using MRO.

```python
class Father:
    def skill(self):
        return "Personality"
class Mother:
    def skill(self):
        return "Beauty"
class Child(Father,Mother):
    def skill(self):
        return super().skill() + super().skill()
obj=Child()
print(obj.skill())

# It calls the parent according to the MRO order it only gives father skill as it is in first position
```
