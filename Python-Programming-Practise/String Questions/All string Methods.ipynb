Write a Program to check/validate given String value is Valid Aadhar Number or not.

Input : First line of input consists of a string which represents aadhar number(May include spaces also)
Output : Print whether the given aadhar number is valid or not valid
if valid Print Valid aadhar or else Print Invalid aadhar
Constraints:
Aadhaar card must contains only digits from 0 to 9
Aadhaar number may contains spaces in between every four numbers

```python
s=input()
l=len(s)
if l == 12 or l == 14:
    if l == 12 and s.isdigit():
        print("Valid Aadhaar")
    else:
        if s[0:4].isdigit() and s[4] == " " and s[5:9].isdigit() and s[9] == " " and s[10:].isdigit():
            print("Valid Aadhaar")
        else:
            print("Invalid Aadhaar")
else:
    print("Invalid Aadhaar")
```

password validator with uppercase, lowercase, digits and length(pass)>=8 and also special character rules
Constraints:
Input:             First Line of the Input Consists of String ( Represents Name )
                       Second Line of the Input Consists of String ( Represents Password )
Output:         print Valid Password or Not a Valid Password.
Constraints:
Password Must be consists of 8 Characters
Password Must be Starts with Capital Character
Password Must be consists of atleast one small Character
Password Must be consists of atleast one Special Character
Password Must be consists of atleast one Numerical Value
Password Must not be consists of Given Name ( Either Small or Capital Characters )
If Above constraints is not Satisfy then print Not a Valid Password or else if satisfies print Valid Password.

```python
p = input("Enter password: ")

if (len(p)>=8 and
    any(i.isupper() for i in p) and
    any(i.islower() for i in p) and
    any(i.isdigit() for i in p) and
    any(not i.isalnum() for i in p)):

    print("Valid password")
else:
    print("Invalid password")
```

check whether a given substring exists in a string

```python
s="pineapple"
s1="ppl"
st=0
ed=len(s1)
while ed < len(s):
    if s[st:ed] == s1:
        print(f"{s1} -> start:{st} and end:{ed}")
        break
    st+=1
    ed+=1
```

find the index of a given character in a string

```python
s="tit for tat"
s1="a"
for i in range(0,len(s)):
    if s[i] == s1:
        print(f"{s1} found at {i} index")
```
