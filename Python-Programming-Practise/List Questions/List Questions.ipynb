10. Write a program to calculate the average of odd numbers in a list.

```python
li = [20, 60, 30, 40, 50, 11, 70, 80, 90, 11, 22, 33, 44, 55, 66, 11, 88, 99, 10, 20, 60, 30, 40, 50, 60, 70, 66, 77,88, 99]
su=c=0
for i in range(0,len(li)):
    if li[i]%2==1:
        su+=li[i]
        c+=1
print(su/c)
```

11. Write a program to print all prime numbers present in a list.

```python
import math
def is_prime(n):
    for j in range(2, int(math.sqrt(n)) + 1):
        if n%j==0:
            return False
    return True
li=[22,99,55,66,99,88,2,5,56,46,4,6,16,18,65,53,454,6,854,65]
for i in li:
    if is_prime(i):
        print(i,end=" ")
```

12. Write a program to print the next prime number for each element in the list

```python
def next_prime(n):
    for i in range(n+1,n*n):
        fc=0
        for j in range(2,int(n**0.5)+1):
            if i%j==0:
                fc+=1
        if fc==0:
            return i
l=[22,99,55,66,99,88,2,5,56,46,4,6,16,18,65,53,454,6,854,65]
for i in l:
    print(next_prime(i),end=" ")
```
